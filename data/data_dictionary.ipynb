# Data Dictionary

## Case_Data.csv

| **Attribute** | **Description** |
|---------------|-----------------|
| **Brand**     | The manufacturer or brand name of the case. |
| **Model**     | The model name or number of the case. |
| **Size**      | The physical size of the case, typically described in terms of form factor. (e.g., "ATX", "Micro ATX", "Mini ITX") |
| **Color**     | The primary color of the case. (e.g., "Black", "White") |
| **Price**     | The retail price of the case in Malaysian Ringgit (MYR) |

---

## CPU_Data.csv

| **Attribute** | **Description** |
|---------------|-----------------|
| **Brand**     | The manufacturer or brand name of the CPU. (e.g., "Intel", "AMD") |
| **Name**      | The model name or number of the CPU. (e.g., "i9-12900K", "Ryzen 9 5900X") |
| **Chipset**   | The CPU's chipset or socket compatibility, which defines which motherboards are compatible. (e.g., "LGA 1200", "AM4") |
| **Score**     | A benchmark score that represents the overall performance of the CPU. |
| **Price**     | The retail price of the CPU in Malaysian Ringgit (MYR) |

---

## GPU_Data.csv

| **Attribute**     | **Description** |
|-------------------|-----------------|
| **Brand**         | The manufacturer or brand name of the GPU. (e.g., "NVIDIA", "AMD") |
| **Name**          | The model name or number of the GPU. (e.g., "RTX 3080", "RX 6800 XT") |
| **Score**         | A benchmark score that represents the overall performance of the GPU. |
| **Series**        | The series  of the GPU (e.g., RTX4080, GTX1050) |
| **Price**         | The retail price of the GPU in Malaysian Ringgit (MYR) |
| **Power Consumption** | The amount of power (in watts) the GPU requires to operate. |
| **Recommended Power** | The recommended total system power supply wattage for the GPU to function optimally. |

---

## MOBO_Data.csv

| **Attribute**     | **Description** |
|-------------------|-----------------|
| **Brand**         | The manufacturer or brand name of the motherboard. |
| **Name**          | The model name or number of the motherboard. |
| **Socket**        | The CPU socket type compatible with the motherboard. (e.g., "LGA 1200", "AM4") |
| **Size**          | The size of the motherboard, typically described as its form factor. (e.g., "ATX", "Micro ATX", "Mini ITX") |
| **RAM Slot**      | The number of RAM slots available on the motherboard. |
| **NVME Slot**     | The number of NVMe slots available for high-speed SSDs. |
| **Wifi**          | Indicates whether the motherboard has built-in WiFi support (1: Yes, 0: No) |
| **Price**         | The retail price of the motherboard in Malaysian Ringgit (MYR) |

---

## PSU_Data.csv

| **Attribute**     | **Description** |
|-------------------|-----------------|
| **Brand**         | The manufacturer or brand name of the power supply unit (PSU). |
| **Name**          | The model name or number of the PSU. |
| **80+ Rating**    | The efficiency certification for power supplies. It denotes how energy-efficient the PSU is. |
| **Wattage**       | The maximum output power of the PSU in watts. (e.g., "750W", "1000W") |
| **Price**         | The retail price of the PSU in Malaysian Ringgit (MYR) |

## RAM_Data.csv

## Storage_Data.csv

| **Attribute**     | **Description** |
|-------------------|-----------------|
| **Brand**         | The manufacturer or brand name of the storage device. |
| **Model**         | The model name or number of the storage device. |
| **Capacity (GB)** | The storage capacity in gigabytes. |

## top100games.csv

| **Attribute**     | **Description** |
|-------------------|-----------------|
| **app_id**        | A unique identifier for each game on the Steam platform. |
| **name**          | The name of the game as listed on Steam. |
| **recommendations** | The number of positive recommendations or user reviews for the game on Steam. |
| **genres**        | A list of genres that the game falls under. These represent the broad category of the game based on its gameplay. (e.g., "Action", "Adventure", "RPG") |
| **tags**          | A list of tags that describe specific elements or features of the game, such as playstyle or themes. (e.g., "Multiplayer", "Co-op", "Shooter") |
| **CPU**           | The minimum or recommended CPU specifications for the game. |
| **GPU**           | The minimum or recommended GPU specifications for the game. |
| **memory**        | The minimum or recommended amount of RAM required to run the game. |
